```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

sns.set_style('whitegrid')
```

```python
thetas = np.linspace(0, 1, 100)
gs = np.sin(np.pi * thetas)**2

plt.plot(thetas, gs)
plt.plot(thetas, stats.beta(a=2, b=2).pdf(thetas)/1.5)
```
